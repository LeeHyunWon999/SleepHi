```python
# ìˆ«ì duration(ns/us/ms/s/h/min)ì„ ì›ì†Œë³„ 'ë¶„'ìœ¼ë¡œ í™˜ì‚°
def _numeric_duration_to_minutes_elwise(arr: pd.Series) -> pd.Series:
    v = pd.to_numeric(arr, errors="coerce").astype("float").to_numpy()
    out = np.full_like(v, np.nan, dtype="float64")
    a = np.abs(v)

    ns = (a >= 1e11)                             # ì˜ˆ: 1.8e12 â‰ˆ 30ë¶„(ns)
    us = (~ns) & (a >= 1e8)                      # Î¼s
    ms = (~ns) & (~us) & (a >= 1e6)              # ms
    sec= (~ns) & (~us) & (~ms) & (a >= 60)       # s
    minute = (~ns) & (~us) & (~ms) & (~sec) & (a > 0) & (a <= 24*60)
    hour   = (~ns) & (~us) & (~ms) & (~sec) & (~minute) & (a > 0)

    out[ns]     = v[ns]    / (1e9 * 60.0)
    out[us]     = v[us]    / (1e6 * 60.0)
    out[ms]     = v[ms]    / (1e3 * 60.0)
    out[sec]    = v[sec]   / 60.0
    out[minute] = v[minute]
    out[hour]   = v[hour]  * 60.0

    # 0ì€ ìœ íš¨ê°’
    out[np.isfinite(v) & (v == 0)] = 0.0
    return pd.Series(out, index=arr.index)

_hhmm_pat = re.compile(r"^(\d{1,2}):(\d{2})(?::(\d{2})(?:\.(\d+)?)?)?$")

def _safe_to_minutes_scalar(x):
    """ë‹¨ì¼ ê°’ xë¥¼ 'ë¶„(float)'ìœ¼ë¡œ. í•´ì„ ë¶ˆê°€ ì‹œ np.nan (â†’ ì—´ ì „ì²´ ë³€í™˜ ìŠ¤í‚µ íŠ¸ë¦¬ê±°)."""
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return np.nan

    # python datetime/time
    if isinstance(x, dt.time):
        return x.hour*60 + x.minute + x.second/60.0
    if isinstance(x, dt.datetime):
        return x.hour*60 + x.minute + x.second/60.0

    # pandas ê³„ì—´
    if isinstance(x, pd.Timestamp):
        return x.hour*60 + x.minute + x.second/60.0
    if isinstance(x, pd.Timedelta):
        return x.total_seconds()/60.0
    if isinstance(x, np.datetime64):
        try:
            ts = pd.to_datetime(x)
            return ts.hour*60 + ts.minute + ts.second/60.0
        except Exception:
            return np.nan

    # ìˆ«ì: duration ë‹¨ìœ„ ì¶”ì •
    if isinstance(x, (int, float, np.integer, np.floating)):
        if np.isnan(x): 
            return np.nan
        v = float(x); av = abs(v)
        if av >= 1e11: return v/(1e9*60.0) # ns
        if av >= 1e8:  return v/(1e6*60.0) # Î¼s
        if av >= 1e6:  return v/(1e3*60.0) # ms
        if av >= 60:   return v/60.0       # s
        if 0 <= v <= 24*60: return v       # ë¶„
        return v*60.0                       # ì‹œê°„

    # ë¬¸ìì—´: timedelta â†’ datetime â†’ HH:MM(:SS) â†’ ìˆ«ì ë¬¸ìì—´
    if isinstance(x, str):
        sx = x.strip()
        if not sx:
            return np.nan
        try:
            td = pd.to_timedelta(sx)
            if isinstance(td, pd.Timedelta) or hasattr(td, "components"):
                return float(td.total_seconds())/60.0
        except Exception:
            pass
        try:
            ts = pd.to_datetime(sx)
            if isinstance(ts, pd.Timestamp):
                return ts.hour*60 + ts.minute + ts.second/60.0
        except Exception:
            pass
        m = _hhmm_pat.match(sx)
        if m:
            h, m_, s_ = int(m.group(1)), int(m.group(2)), int(m.group(3) or 0)
            return h*60 + m_ + s_/60.0
        try:
            fv = float(sx.replace(",", ""))
            return _safe_to_minutes_scalar(fv)
        except Exception:
            return np.nan

    return np.nan

def series_to_minutes_strict(s: pd.Series) -> (pd.Series, pd.Series):
    """
    ì‹œë¦¬ì¦ˆ ì „ì²´ë¥¼ 'ë¶„'ìœ¼ë¡œ ë³€í™˜.
    - dtypeë³„ ìµœì  ê²½ë¡œ ì‚¬ìš©, object í˜¼í•©ì€ ì›ì†Œë³„ ì•ˆì „ íŒŒì„œ ì‚¬ìš©
    - ë°˜í™˜: (minutes_series, ok_mask) ; ok_mask=Falseê°€ í•˜ë‚˜ë¼ë„ ìˆìœ¼ë©´ ì—´ ì „ì²´ ë³€í™˜ ìŠ¤í‚µ
    """
    if is_timedelta64_dtype(s):
        out = (s.dt.total_seconds() / 60.0).astype("float")
        ok = s.notna() | (out == 0.0)
        return out, ok

    if is_datetime64_any_dtype(s):
        out = (s.dt.hour*60 + s.dt.minute + s.dt.second/60.0).astype("float")
        ok = s.notna() | (out == 0.0)
        return out, ok

    if is_numeric_dtype(s):
        out = _numeric_duration_to_minutes_elwise(s).astype("float")
        ok = out.notna() | (out == 0.0)
        return out, ok

    # object í˜¼í•©í˜•: ì•ˆì „ íŒŒì„œë¡œ ì›ì†Œë³„ ì²˜ë¦¬
    out = s.map(_safe_to_minutes_scalar).astype("float")
    ok  = out.notna() | (out == 0.0)
    return out, ok

def looks_time_like_series(s: pd.Series, sample_n: int = 200) -> bool:
    """ì—°ì†í˜• í›„ë³´ ì¤‘ 'ì‹œê°„/ì§€ì†ì‹œê°„'ìœ¼ë¡œ ë³´ì´ëŠ”ì§€ íœ´ë¦¬ìŠ¤í‹± íŒì •."""
    if is_datetime64_any_dtype(s) or is_timedelta64_dtype(s):
        return True
    if is_object_dtype(s):
        smp = s.dropna().astype(str).head(sample_n)
        if smp.empty: 
            return False
        if smp.str.contains(r"\bdays\b|^\s*\d{1,2}:\d{2}(:\d{2}(\.\d+)?)?$|^1900-01-01", regex=True).any():
            return True
        has_dt_like = s.dropna().map(lambda v: isinstance(v, (dt.time, dt.datetime, pd.Timestamp, pd.Timedelta, np.datetime64))).any()
        if has_dt_like:
            return True
    if is_numeric_dtype(s):
        smp = s.dropna().astype(float).abs().head(sample_n)
        if smp.empty:
            return False
        if (smp.ge(1e6).mean() > 0.05) or (smp.between(0, 24*60).mean() > 0.5):
            return True
    return False

def convert_time_columns_inplace_for_efa(df: pd.DataFrame, continuous_vars: list):
    """
    - continuous_vars ì¤‘ ì‹œê°„/ì§€ì†ì‹œê°„ìœ¼ë¡œ ë³´ì´ëŠ” ì—´ë§Œ ê³¨ë¼ in-placeë¡œ ë¶„(float)ë¡œ ë³€í™˜
    - í•œ ê°’ì´ë¼ë„ ì‹¤íŒ¨í•˜ë©´ ê·¸ ì—´ì€ ë³€í™˜í•˜ì§€ ì•ŠìŒ(ì›ë³¸ ìœ ì§€)
    - ë°˜í™˜: {'converted': [...], 'skipped': [...], 'issues': DataFrame, 'efa_cols': [...]}
    """
    converted, skipped, issues = [], [], []

    # ì‹œê°„/ì§€ì†ì‹œê°„ í›„ë³´ë§Œ ì„ ë³„
    time_like_candidates = [c for c in continuous_vars if c in df.columns and not is_numeric_dtype(df[c]) and looks_time_like_series(df[c])]

    for col in time_like_candidates:
        minutes, ok = series_to_minutes_strict(df[col])
        if ok.all():
            # ì›ë³¸ ì—´ì„ 'ë¶„(float)'ë¡œ ë®ì–´ì“°ê¸°
            df[col] = minutes.astype("float")
            converted.append(col)
        else:
            skipped.append(col)
            bad = ~ok
            bad_examples = df.loc[bad, col].astype(str).value_counts().head(10)
            issues.append({
                "column": col,
                "n_total": int(len(df)),
                "n_failed": int(bad.sum()),
                "examples": "; ".join([f"{k} x{v}" for k, v in bad_examples.items()])
            })

    # EFAì— ì‹¤ì œë¡œ íˆ¬ì… ê°€ëŠ¥í•œ ì—°ì†í˜• ì—´(ìˆ«ìí˜•ë§Œ)
    efa_cols = [c for c in continuous_vars if c in df.columns and is_numeric_dtype(df[c])]

    issues_df = pd.DataFrame(issues).sort_values("n_failed", ascending=False) if issues else pd.DataFrame(columns=["column","n_total","n_failed","examples"])

    return {
        "converted": converted,
        "skipped": skipped,
        "issues": issues_df,
        "efa_cols": efa_cols
    }

# ===== ì‹¤í–‰ë¶€ =====
# continuous_varsëŠ” ì§ˆë¬¸ì—ì„œ ì´ë¯¸ ë§Œë“¤ì–´ì ¸ ìˆë‹¤ê³  ê°€ì •
report = convert_time_columns_inplace_for_efa(df, continuous_vars)

print("[ë³€í™˜ ì™„ë£Œ ìš”ì•½]")
print("  ë³€í™˜ ì„±ê³µ ì—´ ìˆ˜:", len(report["converted"]))
if report["converted"]:
    print("   â””", report["converted"])

print("  ë³€í™˜ ìŠ¤í‚µ ì—´ ìˆ˜:", len(report["skipped"]))
if report["skipped"]:
    print("   â””", report["skipped"])
    print("\n[ìŠ¤í‚µ ì‚¬ìœ (ë¬¸ì œ ê°’ ì˜ˆì‹œ)]")
    print(report["issues"].to_string(index=False))

# EFA íˆ¬ì… ì—´ ëª©ë¡ (ìˆ«ìí˜•ë§Œ)
efa_cols = report["efa_cols"]
print("\n[EFA ì—´ ê°œìˆ˜]:", len(efa_cols))
# ì´ì œ ë°”ë¡œ EFA:
# X_efa = df[efa_cols].copy()
# print("EFA ì…ë ¥ í¬ê¸°:", X_efa.shape)
```

# Ver2 ì¶”ê°€ : hh:mm íƒ€ì… ê°’ì€ ê° ë³€ìˆ˜ í‰ê· ê°’ê³¼ì˜ ìƒëŒ€ì  ê°’ìœ¼ë¡œ ì „ë¶€ ë³€ê²½

```python
# ---------------------------------------
# hh:mm íƒ€ì… ë³€ìˆ˜ì˜ í‰ê·  ê¸°ì¤€ ìƒëŒ€ê°’ ë³€í™˜
# ---------------------------------------
def apply_relative_to_mean(df, meta):
    # (1) hh:mm íƒ€ì… ë³€ìˆ˜ ëª©ë¡
    cols = meta.loc[meta["íƒ€ì…"].astype(str).str.lower().str.strip() == "hh:mm", "ë³€ìˆ˜ëª…"].dropna().astype(str)
    cols = [c for c in cols if c in df.columns]
    if not cols:
        print("âš ï¸ ë³€í™˜ ëŒ€ìƒ(hh:mm íƒ€ì…) ë³€ìˆ˜ê°€ ì—†ìŠµë‹ˆë‹¤.")
        return df

    df_out = df.copy()
    print(f"\n[ìƒëŒ€ì  ì°¨ì´ ë³€í™˜ ê²°ê³¼ ìš”ì•½] (ë‹¨ìœ„: ë¶„)\n")

    for col in cols:
        s = df_out[col].astype(float)
        mean_val = s.mean(skipna=True)
        df_out[col] = s - mean_val  # í‰ê·  ëŒ€ë¹„ ìƒëŒ€ê°’ìœ¼ë¡œ ì¹˜í™˜

        # ê°„ë‹¨í•œ ì¶œë ¥ (ì›ë³¸ í‰ê· , ìƒ˜í”Œ 3ê°œ ë¹„êµ)
        sample_idx = s.dropna().index[:3]
        sample_vals = s.loc[sample_idx].tolist()
        rel_vals = df_out.loc[sample_idx, col].tolist()

        print(f"ğŸ“˜ ë³€ìˆ˜ëª…: {col}")
        print(f"  ì›ë³¸ í‰ê· : {mean_val:.2f}ë¶„")
        for i, idx in enumerate(sample_idx):
            print(f"   - ìƒ˜í”Œ{i+1}: ì›ë³¸={sample_vals[i]:.2f}, ìƒëŒ€={rel_vals[i]:+.2f}")
        print()

    return df_out


# ---------------------------------------
# ì‹¤í–‰ ì˜ˆì‹œ
# ---------------------------------------
df = apply_relative_to_mean(df, meta)
```

```python
# ë‚˜ëˆ ì§„ ë²”ì£¼ë“¤ì˜ EFA ì í•©ë„ í™•ì¸(KMO, Bartlett)
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity

# 1) ë¶„ì„ì— ì“¸ ë³€ìˆ˜ë§Œ ì¶”ì¶œ -> ì—¬ê¸°ì„œ fa_vars_ ì„ íƒí•  ê²ƒ!!!! ###################
df_fa = df_EFA[fa_vars_all].copy()

# 2) ê²°ì¸¡ì¹˜ ì²˜ë¦¬ (ì—¬ê¸°ì„  ê°„ë‹¨íˆ í–‰ ë‹¨ìœ„ë¡œ ë“œë¡­)
#    ì‹¤ì œë¡  í‰ê· ëŒ€ì²´ / MICE / domain-specific imputing ê³ ë ¤í•´ë„ ë¨
# df_fa = df_fa.dropna() # ì´ë¯¸ ìœ„ì—ì„œ ìˆ˜í–‰í–ˆìŒ

# 3) KMO ê³„ì‚°
kmo_all, kmo_model = calculate_kmo(df_fa)
print("KMO (overall):", round(kmo_model, 3))

# ë³€ìˆ˜ë³„ KMOê°€ ê¶ê¸ˆí•˜ë©´:
kmo_by_var = dict(zip(df_fa.columns, kmo_all))
print("KMO by variable:")
for col, val in kmo_by_var.items():
    print(f"  {col}: {val:.3f}")

# 4) Bartlett êµ¬í˜•ì„± ê²€ì •
chi_square, p_value = calculate_bartlett_sphericity(df_fa)
print("Bartlett's test chi-square:", round(chi_square, 3))
print("Bartlett's test p-value:", p_value)
```

# Ver 3 ì¶”ê°€ : KMO ê²°ê³¼ ê°’ì´ threshold ë¯¸ë§Œì¸ ë³€ìˆ˜ ì³ë‚´ê¸°, ë²”ì£¼ë³„ df ëª¨ë‘ì— ëŒ€í•´ ë™ì‘

```python
from factor_analyzer.factor_analyzer import calculate_kmo

# EFA ì „ì²˜ë¦¬ ê°€ì ¸ì˜´
def preprocess_for_fa(df, fa_vars, min_var=1e-12):
    """EFAìš© ì „ì²˜ë¦¬: ìˆ«ìí™”â†’Inf/NaN ì²˜ë¦¬â†’ì „ë¶€ ê²°ì¸¡ ì—´ ì œê±°â†’ì¤‘ì•™ê°’ ëŒ€ì¹˜â†’ë¶„ì‚°â‰ˆ0 ì—´ ì œê±°"""
    X = df[fa_vars].copy()
    X = X.apply(pd.to_numeric, errors='coerce')
    X = X.replace([np.inf, -np.inf], np.nan)

    # ì „ë¶€ ê²°ì¸¡ ì—´ ì œê±°
    all_nan_cols = X.columns[X.isna().all()]
    if len(all_nan_cols) > 0:
        X = X.drop(columns=all_nan_cols)

    # ì¤‘ì•™ê°’ ëŒ€ì¹˜
    imp = SimpleImputer(strategy='median')
    X_imp = pd.DataFrame(imp.fit_transform(X), columns=X.columns, index=X.index)

    # ë¶„ì‚°â‰ˆ0 ì—´ ì œê±°
    stds = X_imp.std(ddof=0)
    keep = stds > min_var
    X_imp = X_imp.loc[:, keep]

    # ë‚¨ì€ ì—´ì´ ìµœì†Œ 3ê°œ ë¯¸ë§Œì´ë©´ EFA ë¶ˆê°€
    if X_imp.shape[1] < 3:
        raise ValueError(f"[ì¤‘ë‹¨] ì •ì œ í›„ ë³€ìˆ˜ ìˆ˜ ë¶€ì¡±: {list(X_imp.columns)}")

    return X_imp

def drop_low_kmo_vars(df: pd.DataFrame,
                      kmo_thr: float = 0.50,
                      verbose: bool = True):
    """
    ê°œë³„ KMOê°€ kmo_thrë³´ë‹¤ ì‘ì€ ë³€ìˆ˜ë“¤ì„ ì œê±°.
    dfëŠ” ì´ë¯¸ ìˆ«ìí™”/ê²°ì¸¡ëŒ€ì¹˜/ë¶„ì‚°ì œê±°ê°€ ëë‚œ ìƒíƒœì—¬ì•¼ í•¨.
    """
    kmo_all, kmo_model = calculate_kmo(df)
    kmo_by_var = dict(zip(df.columns, kmo_all))

    removed_vars = [col for col, val in kmo_by_var.items() if val < kmo_thr]
    kept_vars    = [col for col in df.columns if col not in removed_vars]

    if verbose:
        print(f"  - KMO overall: {kmo_model:.3f}")
        print(f"  - threshold : {kmo_thr}")
        if removed_vars:
            print("  - removed by KMO:")
            for rv in removed_vars:
                print(f"      {rv}: {kmo_by_var[rv]:.3f}")
        else:
            print("  - removed by KMO: (none)")

    return df[kept_vars].copy(), removed_vars, kmo_by_var, kmo_model



###### ì„ê³„ê°’ ###########
###### 0ìœ¼ë¡œ ì§€ì •í•˜ë©´ KMO í•„í„°ë§ ì•ˆ í•¨ ################
kmo_thr = 0

###### ì„ì‹œ : ì „ì²´ ë°ì´í„° ëŒ€ìƒìœ¼ë¡œ KMO ìˆ˜í–‰í•˜ê¸° ################
# kmo_blocks = fa_blocks
kmo_blocks = {"all":fa_vars_all} # ì „ì²´ ëŒ€ìƒ ì§€ì •(ì•¼ë§¤)



# ê²°ê³¼ ë‹´ì„ ê·¸ë¦‡
kmo_filtered_blocks = {}     # prefix â†’ (ì •ì œëœ df)
kmo_removed_vars = {}        # prefix â†’ [ì œê±°ëœ ë³€ìˆ˜ë“¤]
kmo_by_var_all = {}          # prefix â†’ {var: kmo}
kmo_overall_all = {}         # prefix â†’ overall kmo

for prefix, var_list in kmo_blocks.items():
    print(f"\n[{prefix}] KMO filtering start...")

    # 1) ì¼ë‹¨ ìˆ˜ì¹˜í™”/ê²°ì¸¡ëŒ€ì¹˜/ë¶„ì‚°â‰ˆ0 ì œê±°
    try:
        X_imp = preprocess_for_fa(df_EFA, var_list)
    except ValueError as e:
        # ë³€ìˆ˜ 3ê°œ ë¯¸ë§Œ ë“±ìœ¼ë¡œ preprocess ë‹¨ê³„ì—ì„œ ë§‰íˆëŠ” ì¼€ì´ìŠ¤
        print(f"  ! skipped: {e}")
        continue

    # 2) KMO ê¸°ì¤€ìœ¼ë¡œ ë³€ìˆ˜ ë–¨êµ¬ê¸°
    X_kmo, removed_vars, kmo_by_var, kmo_overall = drop_low_kmo_vars(
        X_imp,
        kmo_thr=kmo_thr,
        verbose=True
    )

    # 3) ë³€ìˆ˜ 3ê°œ ë¯¸ë§Œì´ë©´ ì´ ë²”ì£¼ëŠ” EFA ì•ˆ ëŒë¦¬ëŠ” ê²Œ ë‚«ë‹¤
    if X_kmo.shape[1] < 3:
        print(f"  ! after KMO filtering, vars={X_kmo.shape[1]} < 3 â†’ skip EFA for [{prefix}]")
    else:
        print(f"  - kept vars ({X_kmo.shape[1]}): {list(X_kmo.columns)}")

    # 4) ê²°ê³¼ ì €ì¥
    kmo_filtered_blocks[prefix] = X_kmo
    kmo_removed_vars[prefix] = removed_vars
    kmo_by_var_all[prefix] = kmo_by_var
    kmo_overall_all[prefix] = kmo_overall

print("\n=== KMO filtering summary ===")
for prefix in kmo_blocks.keys():
    if prefix not in kmo_filtered_blocks:
        continue
    print(f"- {prefix}: overall KMO={kmo_overall_all[prefix]:.3f}, "
          f"kept={kmo_filtered_blocks[prefix].shape[1]}, "
          f"removed={len(kmo_removed_vars[prefix])}")
```

# ì´ì œ EFA ì‹œì‘ : í•˜ë‚˜ì— ëŒ€í•´ ë¶„ì„í•˜ëŠ” ìš©ë„
## Scree Plotë¶€í„° ë½‘ê¸°

```python
# ë³€ìˆ˜ ì„ ì •

fa_vars_kmo_removed_list = {k: list(v.columns) for k, v in kmo_filtered_blocks.items()}

###### ì´ê±° ê±´ë“œë¦¬ë©´ì„œ í•˜ë‚˜ì”© EFAí•  ê²ƒ #########################
fa_vars = fa_vars_kmo_removed_list["all"]


if len(fa_vars) < 3:
    raise ValueError(f"[skip] ìš”ì¸ë¶„ì„ ë³€ìˆ˜ ìˆ˜ ë¶€ì¡±: {fa_vars}")

# --- ì •ì œ ë‹¨ê³„ ---
X = df_EFA[fa_vars].copy()
X = X.apply(pd.to_numeric, errors='coerce')
X = X.replace([np.inf, -np.inf], np.nan)

# 1) ì „ë¶€ ê²°ì¸¡ ì—´ í™•ì¸
all_nan_cols = X.columns[X.isna().all()]
if len(all_nan_cols) > 0:
    print("[ì œê±°] ì „ë¶€ ê²°ì¸¡ì¸ ì—´:", list(all_nan_cols))
X = X.drop(columns=all_nan_cols)

# 2) ì¤‘ì•™ê°’ ëŒ€ì¹˜
imp = SimpleImputer(strategy='median')
X_imp = pd.DataFrame(imp.fit_transform(X), columns=X.columns, index=X.index)

# 3) ë¶„ì‚° 0(ë˜ëŠ” ê·¹ì†Œ) ì—´ í™•ì¸
stds = X_imp.std(ddof=0)
zero_var_cols = stds[stds <= 1e-5].index
if len(zero_var_cols) > 0:
    print("[ì œê±°] ë¶„ì‚°â‰ˆ0ì¸ ì—´:", list(zero_var_cols))
X_imp = X_imp.drop(columns=zero_var_cols)

# 4) ìƒê´€í–‰ë ¬ NaN ë°œìƒ ì—¬ë¶€ ì ê²€
corr = X_imp.corr()
bad_in_corr = corr.columns[corr.isna().any()].tolist()
if len(bad_in_corr) > 0:
    print("[ê²½ê³ ] ìƒê´€í–‰ë ¬ì—ì„œ NaN ë°œìƒí•œ ì—´:", bad_in_corr)

# 5) ë³€ìˆ˜ ìˆ˜ ìµœì¢… í™•ì¸
if X_imp.shape[1] < 3:
    raise ValueError(f"[ì¤‘ë‹¨] ì •ì œ í›„ ë³€ìˆ˜ ìˆ˜ ë¶€ì¡±: {list(X_imp.columns)}")

# --- ìŠ¤í¬ë¦¬ í”Œë¡¯ ---
fa_scree = FactorAnalyzer(rotation='oblimin')
fa_scree.fit(X_imp)

eigenvals, _ = fa_scree.get_eigenvalues()
print("[EFA] ê³ ìœ ê°’(Eigenvalues):", np.round(eigenvals[:X_imp.shape[1]], 4))

plt.figure()
plt.plot(range(1, len(eigenvals) + 1), eigenvals, marker='o')
plt.axhline(1, linestyle='--')
plt.xlabel('Factor')
plt.ylabel('Eigenvalue')
plt.title('Scree Plot')
plt.show()
```

```python
# EFA ì‹œì‘, ì € ë°‘ì— N_FACTORS ì„¤ì • í›„ ì‹¤í–‰!!!!!

def preprocess_for_fa(df, fa_vars, min_var=1e-12):
    """EFAìš© ì „ì²˜ë¦¬: ìˆ«ìí™”â†’Inf/NaN ì²˜ë¦¬â†’ì „ë¶€ ê²°ì¸¡ ì—´ ì œê±°â†’ì¤‘ì•™ê°’ ëŒ€ì¹˜â†’ë¶„ì‚°â‰ˆ0 ì—´ ì œê±°"""
    X = df[fa_vars].copy()
    X = X.apply(pd.to_numeric, errors='coerce')
    X = X.replace([np.inf, -np.inf], np.nan)

    # ì „ë¶€ ê²°ì¸¡ ì—´ ì œê±°
    all_nan_cols = X.columns[X.isna().all()]
    if len(all_nan_cols) > 0:
        X = X.drop(columns=all_nan_cols)

    # ì¤‘ì•™ê°’ ëŒ€ì¹˜
    imp = SimpleImputer(strategy='median')
    X_imp = pd.DataFrame(imp.fit_transform(X), columns=X.columns, index=X.index)

    # ë¶„ì‚°â‰ˆ0 ì—´ ì œê±°
    stds = X_imp.std(ddof=0)
    keep = stds > min_var
    X_imp = X_imp.loc[:, keep]

    # ë‚¨ì€ ì—´ì´ ìµœì†Œ 3ê°œ ë¯¸ë§Œì´ë©´ EFA ë¶ˆê°€
    if X_imp.shape[1] < 3:
        raise ValueError(f"[ì¤‘ë‹¨] ì •ì œ í›„ ë³€ìˆ˜ ìˆ˜ ë¶€ì¡±: {list(X_imp.columns)}")

    return X_imp

# ==== ì‚¬ìš©ìê°€ ì„ íƒí•œ ìš”ì¸ ìˆ˜ ====
N_FACTORS = 8  # <- ìŠ¤í¬ë¦¬ í”Œë¡¯ ë³´ê³  ê²°ì •

# fa_vars ëŠ” ê¸°ì¡´ê³¼ ë™ì¼í•˜ê²Œ ì •ì˜ë˜ì–´ ìˆë‹¤ê³  ê°€ì •
X_imp = preprocess_for_fa(df_EFA, fa_vars)   # âœ… ìŠ¤í¬ë¦¬í”Œë¡¯ê³¼ ë™ì¼ ì „ì²˜ë¦¬ ì ìš©

# (ì„ íƒ) í‘œì¤€í™”: ìŠ¤ì¼€ì¼ ì°¨ì´ê°€ í¬ë©´ ê¶Œì¥
# X_imp = (X_imp - X_imp.mean()) / X_imp.std(ddof=0)

# N_FACTORS ê²€ì¦: ë³€ìˆ˜ ìˆ˜ë¥¼ ë„˜ì„ ìˆ˜ ì—†ìŒ
max_factors = min(X_imp.shape[1]-1, X_imp.shape[0]-1)  # ë³´ìˆ˜ì ìœ¼ë¡œ ì œí•œ
if N_FACTORS > max_factors:
    raise ValueError(f"N_FACTORS={N_FACTORS}ê°€ í—ˆìš©ì¹˜({max_factors})ë¥¼ ì´ˆê³¼")

# 1-ìš”ì¸ì¼ ë•Œ varimax íšŒì „ì€ ì˜ë¯¸ê°€ ê±°ì˜ ì—†ìŒ(keep ê·¸ëŒ€ë¡œ ì‚¬ìš©í•´ë„ ë¬´ë°©)
fa = FactorAnalyzer(n_factors=N_FACTORS, rotation="promax")
fa.fit(X_imp)

# ì ì¬ì¹˜ í–‰ë ¬
cols = [f"Factor{i+1}" for i in range(N_FACTORS)]
loadings = pd.DataFrame(fa.loadings_, index=X_imp.columns, columns=cols)
print(loadings.round(3).to_string())
loadings.round(3).to_csv("factor_loadings.csv", index=True, encoding="utf-8-sig") # íŒŒì¼ ì €ì¥

# # ---- Heatmap (ì¶• ê°œìˆ˜ ì œí•œ ì—†ìŒ) ----
# plt.figure(figsize=(1.5*N_FACTORS+4, 0.4*len(loadings)+3))
# sns.heatmap(loadings, annot=True, cmap="coolwarm", center=0, fmt=".2f")
# plt.title("Factor Loadings Heatmap")
# plt.tight_layout()
# plt.show()


# ---- Heatmap (ì¶• ê°œìˆ˜ ì œí•œ ì—†ìŒ, ë‹¨ ë³€ìˆ˜ ë§ìœ¼ë©´ ë‹¨ìˆœí™”) ----
MAX_ANNOT_VARS = 4000  # ì´ ê°’ ë„˜ìœ¼ë©´ ìˆ«ì ì œê±° + í¬ê¸° ê°„ì†Œí™”     ===>>> ì¼ë‹¨ ì—¬ê¸°ì„  ì „ì²´ ê²½í–¥ì„ ë´ì•¼ í•˜ë¯€ë¡œ ê·¸ëƒ¥ ë‹¤ í‘œì‹œ

n_vars = len(loadings)
annot_flag = n_vars <= MAX_ANNOT_VARS

fig_height = min(0.4 * n_vars + 3, 25)  # ì„¸ë¡œì¶• ìµœëŒ€ í¬ê¸° ì œí•œ (ì˜ˆ: 25ì¸ì¹˜)

plt.figure(figsize=(1.5*N_FACTORS+4, fig_height))
sns.heatmap(
    loadings,
    annot=annot_flag,
    cmap="coolwarm",
    center=0,
    fmt=".2f" if annot_flag else "",
    cbar=True
)
plt.title("Factor Loadings Heatmap" + ("" if annot_flag else " (values hidden)"))
plt.tight_layout()
plt.show()



# ---- Loading Scatter: ì• ë‘ ì¶•ë§Œ ----
if N_FACTORS >= 2:
    plt.figure(figsize=(6,6))
    plt.axhline(0, color='gray', linestyle='--')
    plt.axvline(0, color='gray', linestyle='--')
    for var in loadings.index:
        x, y = loadings.loc[var, "Factor1"], loadings.loc[var, "Factor2"]
        plt.scatter(x, y)
        plt.text(x + 0.02, y, var)
    plt.xlabel("Factor1")
    plt.ylabel("Factor2")
    plt.title("Factor Loadings (First Two Factors)")
    plt.tight_layout()
    plt.show()
else:
    print("[info] N_FACTORS=1: ì‚°ì ë„(2D) ìƒëµ")
```

# Ver 2 ì¶”ê°€ : CFAë¥¼ ìœ„í•œ ìš”ì¸êµ¬ì¡° ì •ì œ ì „ì²˜ë¦¬
# Ver 3 ë³€ê²½ í•„ìš” : ìœ„ì—ì„œ ê³„ì¸µì ìœ¼ë¡œ í˜•ì„±í•œ ë°ì´í„° ê¸°ë°˜ìœ¼ë¡œ ìˆ˜ì • ë‚´ì§€ ì¶”ê°€ ì •ì œí•˜ëŠ” ì½”ë“œë¡œ ë³€ê²½í•´ì•¼ í•¨
- ê³µí†µì„±ì´ íŠ¹ì • ê°’ ì´í•˜ì¸ ë³€ìˆ˜ ì œê±° (ê¸°ë³¸ : 0.2)
- ê°€ì¥ ë†’ì€ ì ì¬ëŸ‰(ì ˆëŒ€ê°’)ì´ íŠ¹ì • ê°’ ì´í•˜ì¸ ë³€ìˆ˜ ì œê±° (ê¸°ë³¸ : 0.35)
- ì ì¬ëŸ‰ì´ ê°€ì¥ ë†’ì€ ìš”ì¸ìœ¼ë¡œ êµ¬ì¡°í™”

```python
from itertools import zip_longest
import pandas as pd

def refine_for_cfa(
    result,
    loadings,
    n_factors,
    communality_thr=0.5,
    loading_thr=0.5,
    cross_loading_thr=0.2  # 1ìˆœìœ„ì™€ 2ìˆœìœ„ loading ì°¨ì´ ê¸°ì¤€
    ### ì—¬ê¸° ê°’ë“¤ì€ ê¸°ë³¸ì„¤ì • ê°’ì´ë¯€ë¡œ ì§ì ‘ ì§€ì •í•˜ëŠ”ê±´ ë°‘ì—ì„œ í•  ê²ƒ!
):
    df_comm = result.copy()
    df_load = loadings.copy()

    # 1ï¸âƒ£ ê³µí†µì„± ê¸°ì¤€
    drop_comm = df_comm[df_comm["Communality"] <= communality_thr].index.tolist()

    # 2ï¸âƒ£ ìµœëŒ€ ì ì¬ëŸ‰ ê¸°ì¤€
    abs_all = df_load.abs()
    max_load = abs_all.max(axis=1)
    drop_low = max_load[max_load <= loading_thr].index.tolist()

    # 3ï¸âƒ£ 1ì°¨ ì œê±° ëŒ€ìƒ (ê³µí†µì„± + ìµœëŒ€ ì ì¬ëŸ‰)
    base_drop_vars = set(drop_comm + drop_low)
    print(f"[ì œê±°] ê³µí†µì„±â‰¤{communality_thr} ë˜ëŠ” ìµœëŒ€ì ì¬ëŸ‰â‰¤{loading_thr} ë³€ìˆ˜ {len(base_drop_vars)}ê°œ:")
    print(base_drop_vars, "\n")

    # 4ï¸âƒ£ 1ì°¨ í•„í„°ë§ í›„ ë°ì´í„°í”„ë ˆì„
    keep_vars = [v for v in df_load.index if v not in base_drop_vars]
    abs_load = abs_all.loc[keep_vars]

    # 5ï¸âƒ£ ìš”ì¸ë³„ ë³€ìˆ˜ êµ¬ì¡°í™” + cross-loading ì²˜ë¦¬
    factor_map = {f"Factor{i}": [] for i in range(1, n_factors + 1)}
    drop_cross = []  # cross-loading ê¸°ì¤€ ë¯¸ë§Œìœ¼ë¡œ ì œê±°ëœ ë³€ìˆ˜ë“¤

    for var in keep_vars:
        row = abs_load.loc[var]

        # loading_thr ì´ìƒì¸ ìš”ì¸ë“¤ë§Œ "ê°•í•œ ì ì¬"ë¡œ ê°„ì£¼
        strong_mask = row >= loading_thr
        strong_factors = row.index[strong_mask].tolist()

        # ê°•í•˜ê²Œ ì ì¬ëœ ìš”ì¸ì´ ì—†ìœ¼ë©´ (ì´ë¡ ìƒ drop_lowì—ì„œ ì´ë¯¸ ì œê±°ë˜ì—ˆìœ¼ë¯€ë¡œ ê±°ì˜ ì•ˆ ë‚˜ì˜¬ ê²ƒ)
        if len(strong_factors) == 0:
            continue

        # â‘  ë‹¨ì¼ ìš”ì¸ì—ë§Œ ê°•í•˜ê²Œ ì ì¬ â†’ ê·¸ëŒ€ë¡œ ê·¸ ìš”ì¸ì— ë°°ì •
        if len(strong_factors) == 1:
            factor_map[strong_factors[0]].append(var)

        # â‘¡ ë‘˜ ì´ìƒì˜ ìš”ì¸ì— ê°•í•˜ê²Œ ì ì¬ëœ ê²½ìš° (cross-loading)
        else:
            # strong_factors ì¤‘ì—ì„œë§Œ ì •ë ¬ (loading ë‚´ë¦¼ì°¨ìˆœ)
            sorted_strong = row[strong_factors].sort_values(ascending=False)
            top1_factor, top1_val = sorted_strong.index[0], sorted_strong.iloc[0]
            top2_val = sorted_strong.iloc[1]  # 2ìˆœìœ„ ê°’

            diff = top1_val - top2_val

            # ì°¨ì´ê°€ cross_loading_thr ì´ìƒì´ë©´ â†’ 1ìˆœìœ„ ìš”ì¸ì—ë§Œ ë°°ì •
            if diff >= cross_loading_thr:
                factor_map[top1_factor].append(var)
            else:
                # ì–´ë–¤ ìš”ì¸ì—ë„ ë°°ì •í•˜ì§€ ì•Šê³  CFAì—ì„œ ì œê±°
                drop_cross.append(var)

    # 6ï¸âƒ£ ìµœì¢… ì œê±° ë³€ìˆ˜ ì§‘í•©
    drop_vars = sorted(base_drop_vars | set(drop_cross))

    print(f"[ì œê±°] êµì°¨ì ì¬(cross-loading) ê¸°ì¤€ ë¯¸ë§Œ(Î”<{cross_loading_thr})ìœ¼ë¡œ ì œê±°ëœ ë³€ìˆ˜ {len(drop_cross)}ê°œ:")
    if drop_cross:
        print(drop_cross)
    print()

    print(f"âœ… ìš”ì¸ë³„ ë³€ìˆ˜ êµ¬ì¡° (cross_loading_thr={cross_loading_thr}, loading_thr={loading_thr})")
    for fname, members in factor_map.items():
        print(f" - {fname} ({len(members)}ê°œ): {members}")

    # 7ï¸âƒ£ ìµœì¢… loadings_filtered (ëª¨ë“  drop_vars ì œê±° í›„)
    df_load_filtered = df_load.drop(index=drop_vars)

    # 8ï¸âƒ£ ê°€ë¡œí˜• CSV ì €ì¥
    factor_names = [f"Factor{i}" for i in range(1, n_factors + 1)]
    factor_lists = [factor_map[f] for f in factor_names]
    df_factors_wide = pd.DataFrame(zip_longest(*factor_lists), columns=factor_names)
    df_factors_wide.to_csv("factor_assignment_wide.csv", index=False, encoding="utf-8-sig")
    print("ğŸ’¾ 'factor_assignment_wide.csv' (ê°€ë¡œí˜•) íŒŒì¼ë¡œ ì €ì¥ ì™„ë£Œ.\n")

    return {
        "loadings_filtered": df_load_filtered,
        "drop_vars": drop_vars,
        "factor_groups": factor_map
    }

cfa_refined = refine_for_cfa(
    result=result,
    loadings=loadings,
    n_factors=N_FACTORS, # ìœ„ì—ì„œ ì„¤ì •ëœ ê°’ìœ¼ë¡œ ì§„í–‰
    communality_thr=0.5,
    loading_thr=0.7,
    cross_loading_thr=0.2  # 1ìœ„ ìš”ì¸ê³¼ 2ìœ„ ìš”ì¸ loading ì°¨ì´ ê¸°ì¤€
)
```

```python
# ----------------------------------------
# CFA êµ¬ì¡°ì‹(desc) ìë™ ìƒì„± (+ ìš”ì¸ë¶„ì‚°=1 ê³ ì • í¬í•¨)
# ----------------------------------------
def make_cfa_description(factor_map):
    """
    factor_map: {'Factor1': ['a','b'], 'Factor2': ['c','d']} í˜•íƒœ
    ë°˜í™˜: lavaan-style formula string
    """
    lines = []
    
    # (1) ìš”ì¸ ì ì¬ì‹ ìƒì„±
    for factor, vars_ in factor_map.items():
        if len(vars_) == 0:
            continue
        joined = " + ".join(vars_)
        line = f"{factor} =~ {joined}"
        lines.append(line)
    
    # (2) ìš”ì¸ë¶„ì‚°=1 ê³ ì • êµ¬ë¬¸ ìë™ ì¶”ê°€
    for factor in factor_map.keys():
        lines.append(f"{factor} ~~ 1*{factor}")
    
    # (3) í•˜ë‚˜ì˜ ë¬¸ìì—´ë¡œ ê²°í•©
    desc = "\n".join(lines)
    
    # (4) ê²°ê³¼ í™•ì¸ìš© ì¶œë ¥
    print("âœ… CFAìš© ìš”ì¸êµ¬ì¡°(desc):\n")
    print(desc)
    
    return desc


# ----------------------------------------
# ì‹¤í–‰ ì˜ˆì‹œ
# ----------------------------------------
desc = make_cfa_description(cfa_refined["factor_groups"])
```

```python
from semopy import Model, calc_stats
import pandas as pd

# ì´ ì½”ë“œë¥¼ ì‹¤í–‰í•˜ê¸° ì „ì— df_CFA (ë°ì´í„°í”„ë ˆì„)ì™€ desc (ëª¨ë¸ êµ¬ì¡° ë¬¸ìì—´)ê°€
# ë©”ëª¨ë¦¬ì— ë¡œë“œë˜ì–´ ìˆì–´ì•¼ í•©ë‹ˆë‹¤.

# --- CFA ëª¨ë¸ë§ ---

# 1. ëª¨ë¸ ì •ì˜
# desc ë³€ìˆ˜ëŠ” ì´ì „ ë‹¨ê³„ì—ì„œ make_cfa_description í•¨ìˆ˜ë¡œ ìƒì„±í–ˆë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤.
# ì˜ˆì‹œ: desc = "Factor1 =~ Var1 + Var2\nFactor2 =~ Var3 + Var4"
model = Model(desc)

# 2. ëª¨ë¸ í•™ìŠµ (í•µì‹¬ ìˆ˜ì • ë¶€ë¶„)
# options={'disp': True}ë¥¼ ì¶”ê°€í•˜ì—¬ ìµœì í™” ê³¼ì •ì˜ ìƒì„¸ ì •ë³´ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
# ê° ë°˜ë³µ(iteration)ë§ˆë‹¤ í˜„ì¬ ìƒíƒœ, í•¨ìˆ˜ ê°’ ë“±ì˜ ì •ë³´ê°€ í™”ë©´ì— ë‚˜íƒ€ë‚©ë‹ˆë‹¤.
print("--- CFA ëª¨ë¸ ìµœì í™”ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤ (ì§„í–‰ ìƒí™© ì¶œë ¥) ---")
model.fit(df_CFA, options={'disp': True})
print("--- CFA ëª¨ë¸ ìµœì í™”ê°€ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤ ---\n")

# 3. ìš”ì¸ ë¶€í•˜ëŸ‰, ì í•©ë„ ì§€í‘œ í™•ì¸ (ê¸°ì¡´ê³¼ ë™ì¼)
print("--- ëª¨ë¸ ë¶„ì„ ê²°ê³¼ ---")
est = model.inspect()
print(est)
# ê²°ê³¼ë¥¼ íŒŒì¼ë¡œ ì €ì¥
try:
    est.to_csv("cfa_estimates.csv", index=False, encoding="utf-8-sig")
    print("\n'cfa_estimates.csv' íŒŒì¼ë¡œ ê²°ê³¼ê°€ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
except Exception as e:
    print(f"\níŒŒì¼ ì €ì¥ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")


print("\n--- ëª¨ë¸ ì í•©ë„ ì§€í‘œ ---")
stats = calc_stats(model)
print(stats)
# ê²°ê³¼ë¥¼ íŒŒì¼ë¡œ ì €ì¥
try:
    stats.to_csv("cfa_fit_stats.csv", index=False, encoding="utf-8-sig")
    print("\n'cfa_fit_stats.csv' íŒŒì¼ë¡œ ê²°ê³¼ê°€ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
except Exception as e:
    print(f"\níŒŒì¼ ì €ì¥ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")

```
